## Code for Face Recognition:

```python
face_cap = cv2.CascadeClassifier("C:/Python310/Lib/site-packages/cv2/data/haarcascade_frontalface_default.xml")
video_cap = cv2.VideoCapture(0)
while True :
    ret, video_data = video_cap.read()
    
    # changing the image to gray color:
    col = cv2.cvtColor(video_data, cv2.COLOR_BGR2GRAY)
    
    # Adding the scales for face:
    faces = face_cap.detectMultiScale(
        col,
        scaleFactor = 1.1,
        minNeighbors = 5, 
        minSize = (30, 30),
        flags = cv2.CASCADE_SCALE_IMAGE
    )
    
    # adding the size for the box:
    for(x,y,w,h) in faces:
        cv2.rectangle(video_data,(x,y),(x+w,y+h), (0,255,0), 2)
    
    # showing the live camera video:
    cv2.imshow("video_live", video_data)
    if cv2.waitKey(1) == ord("a"):
        break
video_cap.release(1)
```
